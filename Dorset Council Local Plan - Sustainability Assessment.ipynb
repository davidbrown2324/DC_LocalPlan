```python


chunks = pd.read_csv('Sustainability_Assessment_Raw.csv', header=None, chunksize=15)

header = ['Sustainability objective', 'Short Term', 'Mid Term', 'Long Term', 'Permanent', 'Comments']

# Create a dictionary to store the data in.
dfs = dict()

#Create a list to save the names of tables.
table_names = []

# Make each chunk a data frame.
for n, chunk in enumerate(chunks):

    # convert to pandas
    df = pd.DataFrame(chunk)
    
    # get table name
    table_name = df.iloc[0, 0]
    
    table_names.append(table_name)
    # lose top three rows
    rows = df.index[[0, 1, 2]]
    df.drop(rows, inplace=True)
    
    # drop the last row which is NaNs
    df.dropna(axis=0, how='all', inplace=True)

    # add header
    df.columns = header
    
    # reset index
    df.reset_index(inplace=True, drop=True)
    
    # Save the data frame to the dictionary, using it's table_name as the key.
    dfs[table_name] = df
```

```python
# Not a complete list
group_codes = ['DEV ', 'ENV ', 'HOUS ', 'ECON ', 'COM ', 
               'CM(BP)', 'CORM', 'UP(BP)', 'BLAND', 'FWPWM', 
               'FDWP', 'FERN', 'STLSTI ', 'SWAN ', 'VER ', 'THLC', 'WARE', 'WIM ', 'ALD', 'LYTM ', 'LM ',
              'STUR ', 'WOL ', 'DORCH ', 'WEY ', 'PORT ', 'CROS ', 'GILL ', 'SHAF ', 'SHER ', 'STAL', 'STNW ', 'BEAM ',
              'BRID ', 'LYME ']
```

```python
group_matches = []

for code in group_codes:

    code_matches = []

    for name in table_names:
        code_matches.append(code in name)
   
    group_matches.append(code_matches)

# Convert to df
grouping = pd.DataFrame(group_matches).T
grouping.columns = group_codes

# Add a column for no matches
grouping['MISC'] = grouping.sum(axis=1) != 1

grouping
```

```python
# Are matches unique?

# Sum all columns
grouping.sum(axis=1).max()

# If max > 1, some code match multiple targets
```

```python
# Collapse wide format into stacked. np.where
groups = pd.DataFrame(
   {'Group': grouping.columns[np.where(grouping)[1]]},
    grouping.index)

groups
```

```python
# Assign groups based on code matching.
group_names = []

for name in table_names:
    for code in group_codes:
        if code in name:
            group_names.append(code)
            break
        else:
            group_names.append('MISC')
        
group_names      
```
